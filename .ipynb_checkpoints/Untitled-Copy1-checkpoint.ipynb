```python
#OLS Model
outcome = 'Weekly_Sales'
x_cols = ['Temperature', 'Fuel_Price', 'CPI', 'Unemployment', 'h_1', 's_2', 's_3',
       's_4', 's_5', 's_6', 's_7', 's_8', 's_9', 's_10', 's_11', 's_12',
       's_13', 's_14', 's_15', 's_16', 's_17', 's_18', 's_19', 's_20', 's_21',
       's_22', 's_23', 's_24', 's_25', 's_26', 's_27', 's_28', 's_29', 's_30',
       's_31', 's_32', 's_33', 's_34', 's_35', 's_36', 's_37', 's_38', 's_39',
       's_40', 's_41', 's_42', 's_43', 's_44', 's_45', 'w_1', 'w_2', 'w_3',
       'w_4', 'w_5', 'w_6', 'm_2', 'm_3', 'm_4', 'm_5', 'm_6', 'm_7', 'm_8',
       'm_9', 'm_10', 'm_11', 'm_12', 'y_2011', 'y_2012']
predictors = '+'.join(x_cols)
formula = outcome + '~' + predictors
model = ols(formula=formula, data=df).fit()
model.summary()
```

```python
import scipy.stats as stats
residuals = model.resid
sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
plt.show()
#plt.savefig("/Users/vineeth/Documents/phase_2/Q-Qplot.png", dpi=150);
```
